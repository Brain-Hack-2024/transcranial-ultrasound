```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Define the function parameters
alpha = 1  # Attenuation in cm^-1 MHz^-1
distance = 1.4  # Bone thickness in cm

# Define the frequency range in MHz
frequencies = np.linspace(0.1, 10, 500)  # Avoid zero to prevent division errors

# Calculate the attenuation
attenuation = 10 * (-alpha * frequencies * distance + 4 * np.log10(frequencies))

# Find the index of the maximum attenuation
max_index = np.argmax(attenuation)
max_frequency = frequencies[max_index]
max_attenuation = attenuation[max_index]

# Plot the function with the maximum point highlighted
# plt.figure(figsize=(8, 6))
plt.plot(frequencies, attenuation, label=r"$10^{-\alpha f} \cdot f^4$")
plt.scatter(
    max_frequency,
    max_attenuation,
    label=f"Max at {max_frequency:.2f} MHz",
    zorder=5,
    marker="o",
)
# plt.yscale("log")  # Use a logarithmic scale for better visualization
plt.xlabel("Frequency (MHz)")
plt.ylabel("Signal Intensity Drop (dB)")
plt.title("Ultrasound Signal Drop through Bone for Blood Cells")
plt.grid(True, which="both", linestyle="--", linewidth=0.5)
plt.legend()
plt.tight_layout()
plt.show()
```
